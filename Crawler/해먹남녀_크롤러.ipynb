# Load Package

```python
from lxml.html import fromstring

import pandas as pd
import numpy as np

import requests
import time
import os
```

# Crawling (1): url

```python
page = 1
df = pd.DataFrame()

while True:
    
    # url 접속
    url = f'https://haemukja.com/recipes?page={page}'
    res = requests.get(url)
    parser = fromstring(res.text)
    
    # 정보수집
    names = [name.text_content() for name in parser.xpath('/html/body/div/div[1]/div[2]/section/div[2]/div/ul/li/p/a/strong')]
    urls = ['https://haemukja.com'+url.get('href') for url in parser.xpath('/html/body/div/div[1]/div[2]/section/div[2]/div/ul/li/p/a')]
    
    # 데이터프레임 추가
    tmp = pd.DataFrame({'레시피':names,
                       'url':urls})
    
    df = pd.concat([df,tmp]).reset_index(drop=True)
    
    # 종료조건
    if len(names) == 0:
        break
        
    # log
    if page % 10 == 0:
        print(f'page: {page}완료 ==')
        
    # page += 1
    page += 1
```

# Crawling (2): 재료

```python
df_final = df.copy()
```

```python
for idx, url in enumerate(df_final['url']):
    
    # url 접속
    res = requests.get(url)
    parser = fromstring(res.text)
    
    # 정보수집
    try:
        cook_time = parser.xpath('/html/body/div/div[1]/div[2]/div/div[1]/section[1]/div/div[1]/dl/dd[1]')[0].text_content()
    except:
        print(f'{idx}번째 url({url})은 존재하지 않습니다.')
        continue
        
    person_num = parser.xpath('/html/body/div/div[1]/div[2]/div/div[1]/section[1]/div/div[3]/div')[0].text_content().replace(' ','').replace('\n','')
    ingredient_names = '|'.join([ingredient.text_content()  for ingredient in parser.xpath('/html/body/div/div[1]/div[2]/div/div[1]/section[1]/div/div[3]/ul/li/span')])
    ingredients_num = '|'.join([ingredient.text_content()  for ingredient in parser.xpath('/html/body/div/div[1]/div[2]/div/div[1]/section[1]/div/div[3]/ul/li/em')])
    nutrients = '|'.join([' '.join(nutrient.text_content().split(' ')[:2]) for nutrient in parser.xpath('/html/body/div/div[1]/div[2]/div/div[1]/section[2]/div[2]/ul/li/p')])
    
    # 데이터프레임 추가
    df_final.loc[idx, '요리시간'] = cook_time
    df_final.loc[idx, '인분'] = person_num
    df_final.loc[idx, '재료명'] = ingredient_names
    df_final.loc[idx, '재료양'] = ingredients_num
    df_final.loc[idx, '영양소'] = nutrients
    
    # log
    if (idx+1) % 10 == 0:
        print(f'{idx+1}개 완료 ==')
```

```python
df_final.to_csv('해먹남녀_크롤링.csv', index=False, encoding='utf-8-sig')
```

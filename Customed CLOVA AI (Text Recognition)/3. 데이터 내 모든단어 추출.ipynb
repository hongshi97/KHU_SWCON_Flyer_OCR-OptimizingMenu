```python
f1 = open('data/gt_train.txt', 'r')
lines1 = f1.readlines()

f2 = open('data/gt_test.txt', 'r')
lines2 = f2.readlines()
```

```python
str_list = list(map(lambda x: x.split('\t')[1].split('\n')[0], lines1))

all_str1 = []
for str_element in str_list:
    str_elements = [i for i in str_element]
    all_str1.extend(str_elements)
    
all_str1 = list(set(all_str1))
```

```python
str_list = list(map(lambda x: x.split('\t')[1].split('\n')[0], lines2))

all_str2 = []
for str_element in str_list:
    str_elements = [i for i in str_element]
    all_str2.extend(str_elements)
    
all_str2 = list(set(all_str2))
```

```python
all_str = list(set(all_str1 + all_str2))
all_str.sort()
''.join(all_str)
```
